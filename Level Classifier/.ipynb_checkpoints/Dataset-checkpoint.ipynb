```python
from PIL import Image

import numpy as np
import glob
```

```python
im = Image.open("images/1-1.png").convert('RGB')
```

```python
im_a = np.array(im)
print(im_a.shape)
im_slice = im_a[np.newaxis, :, 0:240]
```

```python
im_slices = im_a[:, (np.arange(im_a.shape[1] - 240))[:, np.newaxis] + np.arange(240)].transpose(1, 0, 2, 3)
    
print(im_slices.shape)
```

```python
display(Image.fromarray(im_slices[256]))
```

```python
images = []
for f in glob.iglob("images/?-?.png"):
    print(f)
    images.append(np.array(Image.open(f).convert('RGB')))

images.reverse()
images_arr = np.hstack(images)
display(Image.fromarray(images_arr))
```

```python
image_set = []
for image in images:
    windowed_image = image[:, (np.arange(image.shape[1] - 240))[:, np.newaxis] + np.arange(240)].transpose(1, 0, 2, 3)
    image_set.append(windowed_image)
    print(windowed_image.shape)
```

```python
target_set = [np.full((image_set[i].shape[0], 1), i) for i in range(len(image_set))]
```

```python
image_set_array = np.vstack(image_set)
target_set_array = np.vstack(target_set)
print(image_set_array.shape)
print(target_set_array.shape)
```
